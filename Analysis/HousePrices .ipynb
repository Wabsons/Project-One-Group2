```python
# Dependecies

%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
import os as csv 
import random
 
```

```python
# File to load 

Train_to_load = pd.read_csv('../Resources/train.csv')
```

```python
# Read Data
train_df = (Train_to_load)

train_df.head()
```

```python
#AvgFt_data_df = train_df.loc['SalePrice']
  
filter_saleprice = train_df.filter(items=['SalePrice', 'BedroomAbvGr'])     

Grouped_df = filter_saleprice.groupby('BedroomAbvGr').mean() 

Grouped_df.round()
```

```python
# drop Nan columns
new_data_df = train_df.drop(columns=['Alley', 'PoolQC', 'Fence', 'MiscFeature'])
                                    
new_data_df.head()
```

```python
#Drop NaN columns 
new_train_data_df = new_data_df.dropna() 

new_train_data_df.head()
```

```python
#Retrieve column headings 
list(new_train_data_df) 
```

```python
# Use Series.describe() to get Statistics information
train_df.SalePrice.describe()
```

```python
#calculating the distribution of skewness
plt.hist(train_df.SalePrice, color='blue')
plt.title("Sale Price Distribution")

#save the figure
plt.savefig("SalePriceDistribution.png")

plt.show()
```

```python
#scatter plot  GrLivarea/SalePrice
data = pd.concat([new_train_data_df['SalePrice'], new_train_data_df['GrLivArea']], axis=1)
data.plot.scatter(x='GrLivArea', y= 'SalePrice', ylim=(0,800000),color='red')
plt.title("GrLivArea vs. Sale Price")

#save the figure
plt.savefig("GrLivAreavsSalePrice.png")
```

```python
#scatter plot BedroomAbvGr/SalePrice
data = pd.concat([new_train_data_df['SalePrice'], new_train_data_df['BedroomAbvGr']], axis=1)
data.plot.scatter(x='BedroomAbvGr', y= 'SalePrice', ylim=(0,800000))
plt.title("BedroomAbvGr vs. Sale Price")

#save the figure
plt.savefig("BedroomAbvGrvsSalePrice.png")
```

```python
#scatter plot TotalBsmtSF/SalePrice

data = pd.concat([new_train_data_df['SalePrice'], new_train_data_df['TotalBsmtSF']], axis=1)
data.plot.scatter(x='TotalBsmtSF', y= 'SalePrice', ylim=(0,800000),color='g')
plt.title("TotalBsmtSF vs. Sale Price")

#save the figure
plt.savefig("TotalBsmtSFvsSalePrice.png")
```

```python
#looking at SaleCondition by constructing and plotting a pivot table

condition_pivot = new_train_data_df.pivot_table(index='SaleCondition', values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='yellow')
plt.xlabel('Sale Condition')
plt.ylabel('Sale Price')
plt.xticks(rotation=0)
plt.title("Sale Condition vs. Sale Price")

#save the figure
plt.savefig("SaleConditionvsSalePrice.png")

# Show the plot
plt.show()
```

```python
#looking at lotconfig by constructing and plotting a pivot table

lot_pivot = new_train_data_df.pivot_table(index='LotConfig', values='SalePrice', aggfunc=np.median)
lot_pivot.plot(kind='bar', color='purple')
plt.xlabel('LotConfig')
plt.ylabel('Sale Price')
plt.xticks(rotation=0)
plt.title("LotConfig vs. Sale Price")

#save the figure
plt.savefig("LotConfigvsSalePrice.png")

#show the plot
plt.show()
```
